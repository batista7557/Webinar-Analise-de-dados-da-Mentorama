Importar bibliotecas

```python
#importar pandas para processamento de dados
import pandas as pd
```

```python
#importar biblioteca para gerar gráficos
import matplotlib.pyplot as plt
```

Dados coletados do: https://www.kaggle.com/arjunprasadsarkhel/2021-olympics-in-tokyo

```python
#Biblioteca pandas armazena os valores da tabela 'Medals' numa variável 
medalhas = pd.read_excel('/content/Medals.xlsx')
```

```python
#Exibição dos valores da tabela na variável
medalhas
```

```python
#Exibição dos 10 primeiros valores
medalhas.head(10)
```

```python
#Gráfico gerado dos 10 primeiros países com maior número de medalhas
topGeral = medalhas.sort_values('Total', ascending=False).head(10)
plt.bar(topGeral['Team/NOC'], topGeral['Total'])
plt.xticks(rotation=90)
plt.show()
```

```python
#Gráfico gerado dos 10 primeiros países com maior número de medalhas de ouro
top10ouro = medalhas.sort_values('Gold', ascending=False).head(10)
plt.bar(top10ouro['Team/NOC'], top10ouro['Gold'])
plt.xticks(rotation=90)
plt.show()
```

```python
#Gráfico gerado dos 10 primeiros países com maior número de medalhas de prata
top10prata = medalhas.sort_values('Silver', ascending=False).head(10)
plt.bar(top10prata['Team/NOC'], top10prata['Silver'])
plt.xticks(rotation=90)
plt.show()
```

```python
#Gráfico gerado dos 10 primeiros países com maior número de medalhas de bronze
top10bronze = medalhas.sort_values('Bronze', ascending=False).head(10)
plt.bar(top10bronze['Team/NOC'], top10bronze['Bronze'])
plt.xticks(rotation=90)
plt.show()
```
